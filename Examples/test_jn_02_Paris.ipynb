```python
from tessellation import Tessellation
```

# Create Tessllation Object

```python
paris= Tessellation('Paris')
```

```python
paris.get_polygon().plot().set_axis_off();
```

# Squares and Hexagons

```python
paris_hex=paris.hexagons(7)
paris_hex.plot(lw=1, edgecolor='w').set_axis_off();
```

```python
paris_sqr=paris.squares(14)
paris_sqr.plot(lw=1, edgecolor='w').set_axis_off();
```

# Voronoi

```python
paris_voronoi= paris.voronoi(cluster_algo='k-means',
                        poi_categories=['amenity', 'shop'],
                        timeout=60,
                        n_polygons=500,
                        verbose=True)
```

```python
paris_voronoi.plot(lw=1, edgecolor='w').set_axis_off();
print(len(paris_voronoi), 'polygons')
```
